<p style="color:#64748b;font-size:0.75em;letter-spacing:2px;text-transform:uppercase;margin:16px 0 2px 0">Step 01 &nbsp;—&nbsp; Methodology</p>

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(15, 6.2))
fig.patch.set_facecolor('#0f172a')
ax.set_facecolor('#0f172a')
ax.set_xlim(0, 15)
ax.set_ylim(0, 6.2)
ax.axis('off')

ax.text(7.5, 5.9, 'Four Isolated AI Sessions — No Circular Validation',
        ha='center', va='center', fontsize=13, fontweight='bold', color='#e2e8f0')
ax.text(7.5, 5.55, 'Each session saw only its own inputs. Firewalled sessions had zero knowledge of prior intent.',
        ha='center', va='center', fontsize=9, color='#64748b')

# card x-centers evenly spaced
xs = [1.8, 5.3, 8.7, 12.2]
CW = 3.1   # card width
CH = 4.8   # card height
CY = 0.5   # card bottom y

roles = [
    ('#3b82f6', '#0c1a33', '#1e3a5f', '🔨', 'BUILDER',   'Session 1', False,
     'Wrote python-chess board\nsimulation from PGN files',
     [('Output', True, '#3b82f6'),
      ('extract_features.py', False, '#94a3b8'),
      ('6 positional features', False, '#94a3b8'),
      ('snapshots at ply 20 / 30 / 40–60', False, '#94a3b8')]),

    ('#ef4444', '#330c0c', '#5f1e1e', '🔍', 'CRITIC',    'Session 2  ·  Firewalled', True,
     'Reviewed code output only —\nnever saw Builder intent or prompts',
     [('Caught', True, '#ef4444'),
      ('✗  check_rate confounded', False, '#fca5a5'),
      ('     with material advantage', False, '#64748b'),
      ('✗  tripled pawns counted as 1', False, '#fca5a5')]),

    ('#10b981', '#0a2e1e', '#1e5f3a', '📊', 'ANALYST',   'Session 3', False,
     "Mapped Cohen's d effect sizes\nto engine weight multipliers",
     [('Key insight', True, '#10b981'),
      ('Diagnosed color baseline confound', False, '#94a3b8'),
      ('d = 0.09  →  0.71 after fix', False, '#6ee7b7'),
      ('Designed fixed threshold rule', False, '#94a3b8')]),

    ('#a855f7', '#1a0a33', '#3a1e5f', '⚖️',  'JUDGE',     'Session 4  ·  Firewalled', True,
     'Scored 3 pipelines blind against\na standardised 25-point rubric',
     [('Scores', True, '#a855f7'),
      ('Pipeline A (ours):  19 / 25  🥇', False, '#c4b5fd'),
      ('Pipeline B:           12 / 25  🥈', False, '#94a3b8'),
      ('Pipeline C:           10 / 25  🥉', False, '#94a3b8')]),
]

for i, (x, (color, bg_d, bg_h, icon, title, sub, fw, desc, outputs)) in enumerate(zip(xs, roles)):
    # Main card
    ax.add_patch(mpatches.FancyBboxPatch(
        (x - CW/2, CY), CW, CH,
        boxstyle='round,pad=0.12', facecolor=bg_d,
        edgecolor=color, linewidth=1.8, zorder=2))
    # Header band
    ax.add_patch(mpatches.FancyBboxPatch(
        (x - CW/2, CY + CH - 1.2), CW, 1.2,
        boxstyle='round,pad=0.06', facecolor=bg_h,
        edgecolor='none', zorder=3))

    top = CY + CH
    ax.text(x, top - 0.35, icon,  ha='center', va='center', fontsize=22, zorder=4)
    ax.text(x, top - 0.78, title, ha='center', va='center', fontsize=11,
            fontweight='bold', color=color, zorder=4)
    ax.text(x, top - 1.05, sub,   ha='center', va='center', fontsize=7,
            color='#64748b', style='italic', zorder=4)

    # Desc
    sep_y = top - 1.22
    ax.plot([x - CW/2 + 0.2, x + CW/2 - 0.2], [sep_y, sep_y],
            color=color, lw=0.5, alpha=0.3, zorder=3)
    ax.text(x, sep_y - 0.25, desc, ha='center', va='top',
            fontsize=8, color='#cbd5e1', zorder=4, linespacing=1.55,
            multialignment='center')

    # Output block
    out_y = sep_y - 0.25 - (desc.count('\n') + 1) * 0.28 - 0.22
    ax.plot([x - CW/2 + 0.2, x + CW/2 - 0.2], [out_y, out_y],
            color=color, lw=0.5, alpha=0.2, zorder=3)
    for j, (txt, bold, col) in enumerate(outputs):
        ax.text(x - CW/2 + 0.3, out_y - 0.18 - j * 0.36, txt,
                ha='left', va='center', fontsize=7.8 if bold else 7.5,
                color=col, fontweight='bold' if bold else 'normal',
                fontfamily='monospace' if (not bold and txt.startswith((' ', '✗', 'd =', 'P', 'e', 's', '6'))) else 'sans-serif',
                zorder=4)

    # Firewall badge
    if fw:
        ax.add_patch(mpatches.FancyBboxPatch(
            (x - 1.1, CY + 0.08), 2.2, 0.36,
            boxstyle='round,pad=0.04', facecolor='#1c0606',
            edgecolor='#dc2626', linewidth=1.1, zorder=4))
        ax.text(x, CY + 0.26, '🔒  FIREWALLED',
                ha='center', va='center', fontsize=8,
                color='#f87171', fontweight='bold', zorder=5)

# Arrows
for x1, x2 in zip([x + CW/2 for x in xs[:-1]], [x - CW/2 for x in xs[1:]]):
    mid = (x1 + x2) / 2
    ax.annotate('', xy=(x2, CY + CH/2), xytext=(x1, CY + CH/2),
        arrowprops=dict(arrowstyle='->', color='#475569', lw=2.0))

plt.tight_layout(pad=0.3)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.colors import LinearSegmentedColormap
import numpy as np

TERMS = ['Material', 'King Safety', 'King Attack', 'Mobility',
         'Pawn Structure', 'Piece Activity', 'King Activity']
multipliers = {
    'Karpov':    [1.3, 1.7, 0.7, 1.0, 1.7, 1.0, 0.5],
    'Kasparov':  [1.3, 0.7, 2.0, 1.3, 1.0, 1.7, 0.7],
    'Petrosian': [1.0, 2.0, 0.3, 0.7, 1.7, 0.7, 0.5],
    'Carlsen':   [1.3, 0.5, 1.3, 1.7, 1.3, 1.7, 2.0],
}
source = {
    'Karpov':    ['L','M','L','L','M','L','L'],
    'Kasparov':  ['L','L','M','L','L','M','L'],
    'Petrosian': ['L','M','L','L','L','L','L'],
    'Carlsen':   ['L','M','L','M','L','L','M'],
}
nicknames = {
    'Karpov':    'The Wall · Defensive Positional',
    'Kasparov':  'The Activist · Aggressive Tactical',
    'Petrosian': 'The Theorist · Iron Defence',
    'Carlsen':   'The Grinder · Universal Endgame',
}
df = pd.DataFrame(multipliers, index=TERMS).T

fig = plt.figure(figsize=(15, 5.8))
fig.patch.set_facecolor('#0f172a')
gs = gridspec.GridSpec(1, 2, width_ratios=[2.1, 1], figure=fig,
                       left=0.01, right=0.98, top=0.91, bottom=0.12, wspace=0.06)

# Heatmap
ax1 = fig.add_subplot(gs[0])
ax1.set_facecolor('#0f172a')
cmap = LinearSegmentedColormap.from_list('gm',
    ['#dc2626','#b45309','#1e293b','#059669','#047857'], N=256)
im = ax1.imshow(df.values, cmap=cmap, vmin=0.3, vmax=2.0, aspect='auto')
ax1.set_yticks(range(4))
ax1.set_yticklabels([f"{gm}\n{nicknames[gm]}" for gm in df.index],
                    color='#e2e8f0', fontsize=9.5, fontweight='bold')
ax1.set_xticks(range(7))
ax1.set_xticklabels(TERMS, color='#94a3b8', fontsize=9, rotation=22, ha='right')
ax1.tick_params(length=0)
for sp in ax1.spines.values(): sp.set_visible(False)
for gi, gm in enumerate(df.index):
    for ti, term in enumerate(TERMS):
        val = df.loc[gm, term]
        src = source[gm][ti]
        txt_c = '#0f172a' if 0.55 < val < 1.65 else '#f1f5f9'
        ax1.text(ti, gi, f'{val:.1f}', ha='center', va='center',
                 fontsize=13, fontweight='bold', color=txt_c)
        ax1.text(ti+0.4, gi-0.35, src, ha='center', va='center',
                 fontsize=7.5, fontweight='bold',
                 color='#fca5a5' if src == 'M' else '#334155')
ax1.set_title("Engine Weight Multipliers  (M = Cohen's d measured · L = literature)",
              color='#e2e8f0', fontsize=11, fontweight='bold', pad=10)
cb = plt.colorbar(im, ax=ax1, orientation='horizontal', pad=0.22, fraction=0.03, aspect=40)
cb.set_label('← suppress   neutral   amplify →', color='#64748b', fontsize=7.5)
cb.ax.xaxis.set_tick_params(color='#475569', labelcolor='#64748b', labelsize=7)
cb.outline.set_edgecolor('#334155')

# Radar
ax2 = fig.add_subplot(gs[1], polar=True)
ax2.set_facecolor('#0f172a')
N = len(TERMS)
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist() + [0]
gm_col = {'Karpov':'#3b82f6','Kasparov':'#ef4444','Petrosian':'#10b981','Carlsen':'#f59e0b'}
for gm, row in df.iterrows():
    vals = row.tolist() + [row.tolist()[0]]
    ax2.plot(angles, vals, lw=2.2, color=gm_col[gm], label=gm)
    ax2.fill(angles, vals, alpha=0.1, color=gm_col[gm])
ax2.set_xticks(angles[:-1])
ax2.set_xticklabels(['Mat','KSafe','KAtk','Mob','Pawn','PAct','KAct'],
                    fontsize=8.5, color='#94a3b8')
ax2.set_yticks([0.5, 1.0, 1.5, 2.0])
ax2.set_yticklabels(['0.5','1.0','1.5','2.0'], fontsize=6.5, color='#475569')
ax2.set_ylim(0, 2.3)
ax2.spines['polar'].set_color('#1e293b')
ax2.grid(color='#1e293b', lw=0.8)
ax2.set_title('Style Fingerprints', color='#e2e8f0', fontsize=10.5, fontweight='bold', pad=16)
ax2.legend(loc='upper right', bbox_to_anchor=(1.55, 1.18),
           fontsize=8.5, framealpha=0, labelcolor='#e2e8f0')

import os; os.makedirs('docs', exist_ok=True)
plt.savefig('docs/multiplier_heatmap_final.png', dpi=150, bbox_inches='tight', facecolor='#0f172a')
plt.show()
```

<p style="color:#64748b;font-size:0.75em;letter-spacing:2px;text-transform:uppercase;margin:16px 0 2px 0">Step 03 &nbsp;—&nbsp; Validation</p>

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
from matplotlib.colors import LinearSegmentedColormap
import numpy as np

fig = plt.figure(figsize=(15, 6.8))
fig.patch.set_facecolor('#0f172a')
gs = gridspec.GridSpec(1, 2, width_ratios=[1.05, 1], figure=fig,
                       left=0.03, right=0.97, top=0.91, bottom=0.06, wspace=0.07)

fig.suptitle('Parallel Audit — Blind AI Judge · Standardised 25-pt Rubric',
             color='#475569', fontsize=9, y=0.97)

# ── LEFT: rubric score table (heatmap style) ──────────────────────────────────
ax1 = fig.add_subplot(gs[0])
ax1.set_facecolor('#0f172a')

PIPE_NAMES  = ['A  Our pipeline', 'B  Hand-assigned', 'C  Regex parser']
PIPE_MEDALS = ['🥇', '🥈', '🥉']
PIPE_COLS   = ['#f4c542', '#9ca3af', '#b45309']
CATS        = ['Data Evidence', 'Baseline\nValidity', 'Internal\nConsistency', 'Self-\nCriticism', 'TOTAL']
MAXES       = [10, 5, 5, 5, 25]
RAW_SCORES  = [
    [9, 5, 4, 4, 22],
    [6, 2, 1, 5, 14],
    [0, 0, 4, 3, 7],
]

# Build 3-row x 5-col score matrix (normalised 0-1 for colour)
mat = np.array([[s/m for s, m in zip(row, MAXES)] for row in RAW_SCORES])

cmap = LinearSegmentedColormap.from_list('audit',
    ['#450a0a', '#7c2d12', '#1e293b', '#14532d', '#166534'], N=256)

im = ax1.imshow(mat, cmap=cmap, vmin=0, vmax=1, aspect='auto')

# Axes labels
ax1.set_xticks(range(5))
ax1.set_xticklabels(CATS, color='#94a3b8', fontsize=9, ha='center')
ax1.set_yticks(range(3))
ax1.set_yticklabels([f'{m} {n}' for m, n in zip(PIPE_MEDALS, PIPE_NAMES)],
                    color='#e2e8f0', fontsize=10, fontweight='bold')
ax1.tick_params(length=0)
for sp in ax1.spines.values(): sp.set_visible(False)

# Score annotations inside cells
for ri, row in enumerate(RAW_SCORES):
    for ci, (sc, mx) in enumerate(zip(row, MAXES)):
        bright = mat[ri, ci] < 0.35 or mat[ri, ci] > 0.72
        txt_c  = '#f1f5f9' if bright else '#0f172a'
        bold   = (ci == 4)  # total column
        ax1.text(ci, ri, f'{sc}/{mx}',
                 ha='center', va='center',
                 fontsize=13 if bold else 11,
                 fontweight='bold', color=txt_c)

# Column header divider before TOTAL
ax1.axvline(3.5, color='#334155', lw=1.5, ls='--')

ax1.set_title('Rubric Scores by Pipeline',
              color='#e2e8f0', fontsize=11, fontweight='bold', pad=12)

# Max pts row label
for ci, mx in enumerate(MAXES):
    ax1.text(ci, 3.08, f'max {mx}',
             ha='center', va='bottom', fontsize=7, color='#475569')

# ── RIGHT: verdict cards ──────────────────────────────────────────────────────
ax2 = fig.add_subplot(gs[1])
ax2.set_facecolor('#0f172a')
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 6.8)
ax2.axis('off')

ax2.text(5, 6.55, 'Verdict on Each Pipeline',
         ha='center', va='center', fontsize=12, fontweight='bold', color='#e2e8f0')

verdicts = [
    ('🥇  Pipeline A — 19 / 25', '#f4c542', '#1c1a05', [
        ('✅', 'Fixed threshold rule — fully reproducible'),
        ('✅', 'Cross-style baseline eliminates color confound'),
        ('⚠️', 'Mean castle ply ignores variance  (self-flagged)'),
        ('→',  'Post-audit fix raises projected score to ~22 / 25'),
    ]),
    ('🥈  Pipeline B — 12 / 25', '#9ca3af', '#111111', [
        ('✅', 'Highest self-criticism score (5 / 5)'),
        ('❌', 'wins_without_castling used for two opposite claims'),
        ('❌', 'Karpov PGN filter: 2024–25  (peak was 1970–96)'),
        ('❌', 'Karpov King Safety = 3.5×  — no effect size given'),
    ]),
    ('🥉  Pipeline C — 10 / 25', '#b45309', '#1a0e00', [
        ('✅', 'Honest — labeled 0 cells as MEASURED'),
        ('❌', 'Baseline = "average GM" — never actually sampled'),
        ('❌', 'No pooled SD → effect sizes uncomputable'),
        ('❌', 'All multipliers unanchored to real data'),
    ]),
]

CARD_H  = 1.68
CARD_GAP= 0.10
start_y = 6.10

for vi, (title, col, bg, lines) in enumerate(verdicts):
    top = start_y - vi * (CARD_H + CARD_GAP)
    bot = top - CARD_H

    ax2.add_patch(mpatches.FancyBboxPatch(
        (0.2, bot), 9.6, CARD_H,
        boxstyle='round,pad=0.1', facecolor=bg,
        edgecolor=col, linewidth=1.5, zorder=2))

    ax2.text(0.6, top - 0.27, title,
             fontsize=10.5, fontweight='bold', color=col, va='center', zorder=3)

    line_y = top - 0.60
    for icon, text in lines:
        note = icon == '→'
        ax2.text(0.6, line_y, f'{icon}  {text}',
                 fontsize=8.2, va='center', zorder=3,
                 color='#64748b' if note else '#cbd5e1')
        line_y -= 0.275

plt.show()
```
